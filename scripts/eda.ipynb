```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("../data/synthetic_metrics.csv")
df.head()
```

```python
sns.countplot(data=df, x='anomaly_type')
plt.title("Distribution of Anomaly Types")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='anomaly_type', y='load_efficiency')
plt.title("Load Efficiency by Anomaly Type")
plt.xticks(rotation=45)
plt.show()
```
